# Visualizating Data on Exchange Rates

This dataset describes the Euro's (€) daily exchange rates between 1999 and 2021. This dataset was generated by Daria Chemkaeva with data from the European Central Bank. This dataset was downloaded January 2021 and looks at the exchange rate between the € and the American $.

This project goal was to quckly visualize a datset and then pick some factors to investigte and visualize with a focus on telling a story. I selected a time period and noted key points in Brexit and the pound exchange rate.

```python
# Example of how to perform rolling mean 
values = pd.DataFrame()
values['daily_values'] = pd.Series(range(1,20,2))
values

# Calculate mean for each day using  pandas.Series.rolling().mean()
values['rolling_mean_2'] = values['daily_values'].rolling(2).mean()
values

# The number of days used in calculating the rolling window
# also called the moving window

# If the rolling window is 3, calculate the mean between the value 
# of current day and values of previous two days
values['rolling_mean_3'] = values['daily_values'].rolling(3).mean()
values['rolling_mean_5'] = values['daily_values'].rolling(5).mean()
values

# The rolling window can take various values — some common values 
# include seven (one week), 30 (one month), 50 or 100 days, or 365 days (one year)
# As the window gets larger the line becomes smoother 
# This increases the data-ink ratio and can be useful if we want 
# the audience to focus only on long-term trends
```

```python
# Calculate rolling mean for  US_dollar and use a 30 day window 
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar
```

```python
# Calculate rolling mean for US dollar and use a 365 day window 
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(365).mean()
euro_to_dollar

# visualize the rolling mean with 365 day window
plt.plot(euro_to_dollar['Time'], euro_to_dollar['rolling_mean'])
plt.show()
```

# The Euro vs Dollar Before and After Brexit (before and after the vote)

What was the impact of Brexit on the exchange rate of the Euro and UK pound? Let's look at some key time points in the process:
1. June 23: 2016, majority of those who voted on the referendum chose to leave the European Union.
2. March 9th 2017: the UK's Prime Minister formally triggered Article 50 by writing to European Council President Donald Tusk.  This started the 2 year clock for the UK formally leave the EU.
3. March 14, 2019: the UK requested an extension. (March 29, 2019 was the original date the UK was scheduled to leave the EU at 11pm. Additional extensions were requested and granted.)
4. January 31, 2020: the UK left the EU and entered a transition period.
5. December 31, 2020: At 11pm the transition period ended and the United Kingdom exited the EU market and customs union.

```python
# Set up 5 time points covering 2015-2021

# Rename columns
exchange_rates.rename(columns={'[UK pound sterling ]': 'UK_pound', 
                               'Period\\Unit:': 'Time'}, inplace=True)

# Change the Time column to a datetime data type
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])

# Sort the values by Time in ascending order
exchange_rates.sort_values('Time', inplace=True)

# Reset the index (and drop the initial index)
exchange_rates.reset_index(drop=True, inplace=True)

# Assign Time and Us Dollar to a variable
euro_to_pound = exchange_rates[['Time','UK_pound']]

# Check data
euro_to_pound.head(5)
```

```python
# Set up time periods

# Copy columns of interest and set year range 2015 - 2021 for euro exchange rate 
before_after_brexit = euro_to_pound.copy(
                   )[(euro_to_pound['Time'].dt.year >= 2015) & (euro_to_pound['Time'].dt.year <= 2021)]

before_vote = before_after_brexit.copy()[before_after_brexit['Time'].dt.year < 2016]

after_vote = before_after_brexit.copy()[before_after_brexit['Time'].dt.year == 2016]

year_after_vote = before_after_brexit.copy()[before_after_brexit['Time'].dt.year == 2017]

two_year_clock = before_after_brexit.copy(
       )[(before_after_brexit['Time'].dt.year >= 2017) & (before_after_brexit['Time'].dt.year < 2020)]

leave_transition = before_after_brexit.copy()[before_after_brexit['Time'].dt.year == 2020]
                                              
brexited = before_after_brexit.copy()[before_after_brexit['Time'].dt.year == 2021]                                             
```

```python
import matplotlib.pyplot as plt

def make_ticklabels_invisible(fig):
    for i, ax in enumerate(fig.axes):
        ax.text(0.5, 0.5, "ax%d" % (i+1), va="center", ha="center")
        for tl in ax.get_xticklabels() + ax.get_yticklabels():
            tl.set_visible(False)


plt.figure(0)
ax1 = plt.subplot2grid((3,3), (0,0), colspan=3)
ax2 = plt.subplot2grid((3,3), (1,0), colspan=2)
ax3 = plt.subplot2grid((3,3), (1, 2), rowspan=2)
ax4 = plt.subplot2grid((3,3), (2, 0))
ax5 = plt.subplot2grid((3,3), (2, 1))

plt.suptitle("subplot2grid")
make_ticklabels_invisible(plt.gcf())
plt.show()
```
